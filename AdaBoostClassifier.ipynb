```python
import pandas as pd 
df = pd.read_csv('multiple_disease_dataset.csv')
df.head()
```

```python
X = df.drop(columns= ['Disease'])
y = df['Disease']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)
```

```python
from sklearn.ensemble import AdaBoostClassifier 
clf = AdaBoostClassifier().fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
clf1 = AdaBoostClassifier(n_estimators=125, learning_rate=0.5).fit(X_train, y_train)
clf1.score(X_test, y_test)
```

```python
clf2 = AdaBoostClassifier(n_estimators=20, learning_rate=0.05).fit(X_train, y_train)
clf2.score(X_test, y_test)
```

```python
df['Predicted_Disease'] = clf1.predict(X)
```

```python
df.head()
```

```python
df['Accuracy'] = df['Disease'].eq(df['Predicted_Disease'], axis = 0).astype('int') 
```
