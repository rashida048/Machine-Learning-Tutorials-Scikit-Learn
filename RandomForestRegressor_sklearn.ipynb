```python
import pandas as pd 
df = pd.read_csv('housing_data.csv', parse_dates=['date'])
df.head()
```

```python
df['Month'] = df['date'].dt.month
```

```python
df['Year'] = df['date'].dt.year
```

```python
df
```

```python
df.isna().sum()
```

```python
X = df.drop(columns = ['id', 'date', 'price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 64)
```

```python
from sklearn.ensemble import RandomForestRegressor 
reg = RandomForestRegressor()
reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error 
y_pred_train = reg.predict(X_train)
mean_absolute_error(y_train, y_pred_train)
```

```python
y_pred_test = reg.predict(X_test)
mean_absolute_error(y_test, y_pred_test)
```

```python
from sklearn.model_selection import GridSearchCV 
parameters = {'n_estimators': [120, 125, 130], 
             'max_depth': [16, 18, 22],
              'max_features': [9, 10, 11],
              'max_leaf_nodes': [84, 86, 88]
             }
reg1 = RandomForestRegressor()
reg1 = GridSearchCV(reg1, parameters)
```

```python
reg1.fit(X_train, y_train)
```

```python
y_pred_train = reg1.predict(X_train)
mean_absolute_error(y_train, y_pred_train)
```

```python
y_pred_test = reg1.predict(X_test)
mean_absolute_error(y_test, y_pred_test)
```
